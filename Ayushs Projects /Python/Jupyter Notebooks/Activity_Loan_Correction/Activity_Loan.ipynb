```python
from sklearn.utils import resample 
```

```python
# Split the data set for 1 and for 0 
df_major = df_new[df_new["LoanApproved"] == 0]
df_minor = df_new[df_new["LoanApproved"] == 1]
```

```python
df_downsample = resample(df_major, replace=False,n_samples = len(df_minor), random_state=42)
df_downsample.shape
```

```python
df2 = pd.concat([df_minor,df_downsample], axis=0)
df2.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
df2["lb_EducationLevel"] = lb.fit_transform(df2["EducationLevel"])
df2["lb_HomeOwnershipStatus"] = lb.fit_transform(df2["HomeOwnershipStatus"])
```

```python
x = df2.drop(["HomeOwnershipStatus","EducationLevel","LoanApproved"],axis=1)
```

```python
x.columns
```

```python
y = df2["LoanApproved"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
z_test_scaled = scaler.transform(x_test)
```

```python
df_view = pd.DataFrame(x_train_scaled, columns = x_train.columns)
df_view.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter = 1000)
lr.fit(x_train_scaled,y_train)
```

```python
y_pred = lr.predict(z_test_scaled)
```

```python
lr_result = pd.DataFrame({"Predicted":y_pred,"Actual": y_test})
lr_result
```

```python
#Evaluation of the model 
#Confustion metrics divides into 4 parts 

from sklearn.metrics import confusion_matrix
```

```python
cm_lr = confusion_matrix(y_test,y_pred)
print(cm_lr)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)

#The model correclty predicted 783 1s and wrongly 188 1s 
#The model correclty predicted 803 0s and wrongly 138 0s 
```

```python
# Accuracy of the model gives a measure of the overall proportion of correct predictions 
# performance of the negatves and the positives 
# (tp + tn)/(tf+tn+fp+fn)
from sklearn.metrics import accuracy_score
```

```python
lr_accuracy_score = accuracy_score(y_test,y_pred)
print("Accuracy: ", lr_accuracy_score)
```

```python
#Precistion 
# (TP)/(TP + FP)
from sklearn.metrics import precision_score,recall_score
lr_precision = precision_score(y_test,y_pred)
print("Precision: ", lr_precision)
```

```python
##recall 
# how many of the positives are actually positive 
# how many are predicted correctly 
# (TP)/(TP+FN)
lr_recall = recall_score(y_test,y_pred)
print("Recall: ", lr_recall)
```

### Model Performance Summary

After evaluating six different machine learning models for loan approval prediction, several insights emerged. Logistic Regression performed well with 83% accuracy, correctly identifying approved loans 85% of the time when it predicted approval, and capturing 81% of all actual approved loans.

The Random Forest model delivered the strongest overall performance, achieving the highest accuracy at 83.2% and the best recall at 84.9%. This means it was most effective at identifying applicants who should receive loans. Its precision of 82.5% was also solid, indicating a good balance between accuracy and reliability.

Support Vector Machine showed consistent results across all metrics with 82.4% accuracy and balanced precision and recall scores. Gaussian Naive Bayes stood out with the highest precision at 85.5%, meaning it rarely approved loans incorrectly, though it missed more actual approved loans with a lower recall of 72.8%.

The Decision Tree and K-Nearest Neighbors models underperformed compared to the others, with accuracies around 74-76%. This suggests these simpler algorithms may not capture the complex relationships in loan approval data as effectively.

Overall, the Logistic Regression proved to be the most reliable choice for this loan approval prediction task, offering the best combination of accuracy and the ability to correctly identify approved loans.
