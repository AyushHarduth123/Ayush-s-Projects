```python
#ST10365068 
#import csv
data = pd.read_csv("data.csv")
```

```python
#ST10365068 
# Display first few rows 
data.head(2)

```

```python
#ST10365068 
#COunt missing values
print("\nTotal Missing Values:")
print(data.isnull().sum())
```

```python
#ST10365068 
# Fill missing numeric values with the mean
# Had a error with inplace so i just removed it and it worked
round(data['Age'].fillna(data['Age'].mean()),2)
data['Salary'].fillna(data['Salary'].mean())
```

```python
#ST10365068 

#COunt missing values
print("\nTotal Missing Values:")
print(data.isnull().sum())
```

```python
#ST10365068 
# Fill missing categorical values with the mode
# same as before had the same error so i just remove the inplace and it worked 
data['Name'].fillna(data['Name'].mode()[0])
data['Department'].fillna(data['Department'].mode()[0])
```

```python
#ST10365068 
# checking missing values again
#COunt missing values
print("\nTotal Missing Values:")
print(data.isnull().sum())
```

```python
#ST10365068 
# Drop rows where all values are missing
data.dropna(how='all', inplace=True)
```

```python
#ST10365068
# Display the cleaned dataset
print("\nCleaned Dataset:")
print(data)
```
