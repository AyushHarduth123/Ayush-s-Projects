```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams['font.size'] = (12)
```

```python
x = [1,2,5,6,9,12]
y = [2,4,5,6,8,10]
```

```python
#### Basic Line plot 
```

```python
# when u have continuous data that increases over time the line plot can visualise treads over a period of time
# useful for time series 
# represent any variable that is 
# shows the trend in the dataset to show correlation
# its shows its strength in time series problems 


plt.plot(x,y)
plt.title("x versus y")
plt.ylabel("y values")
plt.xlabel("x values")
plt.show()

```

```python
# we can plot multiple lines and multiple features that can be ploted against a label 
# Just remember that the lables of the arrays should be swapped around
y1 = [2,4,6,8,10,13]
y2 = [1,3,5,7,9,11]
y3 = [3,5,7,9,11,13]

plt.plot(x,y1,'b-', label = 'Line 1') #changes the colour of the lines and adds a label for the legend
plt.plot(x,y2,'r--', label = 'Line 2')
plt.plot(x,y3,'g-.', label = 'Line 3')

plt.title("Multiple ys against single x")
plt.ylabel("y values")
plt.xlabel("x values")
plt.legend(loc = "lower right")
plt.show()
```

```python
# Another method is to introduce more parameters 
plt.plot(x,y,
        color = 'red',
        linewidth =2,
        marker = 'o',
        markerfacecolor = 'yellow',
        markeredgecolor = 'red',
        label = 'x plot')

plt.title("sample plot")
plt.legend()
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True, alpha = 0.3)
plt.show()
```

#### Scatter plot 
determines if a relationship exists between variables and can be used to detect patterns, clusters and outliers 

```python
# Scatterr plot 
# visualise between 2 numeric variables that must both be numeric continuous variables 
# determine if their are actual relationships between variables or not 
# if there is no relationship they willl be all over the place 
# Use it to detect outliers 
# can visualise clusters 

plt.scatter(x,y) # key word scatter
plt.title("Scatter Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True, alpha = 0.3)
plt.show()
```

```python
# We can have different types of scatter plots and we willl use synthetic data 
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

colors = np.random.randn(100)
size = 400

plt.scatter(x,y,c=colors,s=size,alpha=0.6,cmap='viridis')
plt.colorbar(label = 'Color Scale')

plt.xlabel("x values")
plt.ylabel("Y vALUES ")

plt.show()
```

#### Bar Plot 
Bar plot is used for categorical data and used to visualise frequency or count of categorical variables. Each of the bars in the bar plot represents a category or class in the categorical variable

```python
categories = ['A','B','C','D','E']
values = [23,45,56,78,32]

plt.bar(categories, values, color='steelblue', edgecolor='black')
plt.title("Bar Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
# Horiszontal bar plot 
categories = ['A','B','C','D','E']
values = [23,45,56,78,32]

plt.barh(categories, values, color='steelblue', edgecolor='black')
plt.title("Horizontal Bar Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
categories = ['Group 1','Group 2','Group 3']
values_1 = [20,35,30]
values_2 = [25,32,34]
values_3 = [18,38,28]

#Location of the pars 
x = np.arange(len(categories))
width = 0.25

plt.bar(x - width, values_1,width,label='series 1',color = 'blue')
plt.bar(x, values_2,width,label='series 2',color = 'green')
plt.bar(x + width, values_1,width,label='series 3',color = 'red')

plt.title("Grouped Bar Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.xticks(x, categories)
plt.legend()

plt.show()
```

#### Histogram
This plot shows distribution of numerical continous data. It groups the data into a range (1-10), (11-20) ... 
all th bars touch each other 


```python
data = np.random.randn(1000)
plt.hist(data, bins=20, density=True, alpha=0.7, edgecolor ='black')
plt.title("Histogram")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.show()
```

#### Multiple plots on the same canvas 



```python
# setting up the canvas as well as the access we would have inside the canvas 
fig, axes = plt.subplots(2,2, figsize = (12,10))
# specify the index of the plot 

axes[0,0].plot([1,2,3],[1,4,9])
axes[0,0].set_title('Line Plot')

axes[0,1].scatter([1,2,3],[1,4,9])
axes[0,1].set_title('Scatter Plot')

axes[1,0].bar(['A','B','C'],[3,7,9])
axes[0,1].set_title('Bar Plot')

axes[1,1].hist(np.random.randn(100),bins = 20)
axes[0,1].set_title('Histogram')

plt.tight_layout()
# prevents one plot to spill into another 
plt.show()
```

```python
from sklearn.datasets import load_iris
```

```python
# this is after cleaning all the data and filling in missing data 
# this is before encoding and scaling 
iris = load_iris()
df = pd.DataFrame(iris.data,columns = iris.feature_names)
df['Species_value'] = iris.target 
df["Species"] = df["Species_value"].map({0:'Setosa',1:'Versicolor', 2: 'Virginica'})
```

#### Univariate Analysis
##### When dealing with cat data 
1. figure out if there is a class imbalance
   Counting the number of unique categoreis and checking for imbalance
   check the distribution
   if there is a class imbalance then apply a imbalance classs like SMOTE

   Understand the meaning of chosing a specific chart

   
   
3. Binary or Multiclass




```python
# Counting to see if there is class imbalance 
print(df['Species'].value_counts())
```

```python
print(f"Class Imbalance ration: {df['Species'].value_counts(normalize=True)}")
```

```python
fig,axes = plt.subplots(1,2,figsize=(12,4))
df['Species'].value_counts().plot(kind='bar', ax=axes[0],title="Class Count")
axes[0].set_ylabel('Frequency')
axes[0].set_xlabel('Species')


df['Species'].value_counts().plot(kind='pie',ax=axes[1], autopct='%1.1f%%',title='Class Proportion')
plt.tight_layout()
plt.show()
```

```python
counts = df['Species'].value_counts()

fig,axes = plt.subplots(1,2,figsize=(12,4))
axes[0].bar(counts.index, counts.values, edgecolor = 'black')
axes[0].set_xlabel("Species")
axes[0].set_ylabel("Count")
axes[0].set_title("Univariate Analysis")



axes[1].pie(counts.values, labels=counts.index,autopct="%1.1f%%")
axes[1].set_title('species Dsitribution')
```

#### Bivariate 
1. Check if the features have a impact on predicting the target 
2. Analysis how each feature is distributed across the target (groups/classes)

```python
df_new = df.drop("Species_value", axis=1)
df_new.head()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df_new, hue='Species', diag_kind='kde', markers=['o','s','D'])# plot the features based on the target 
plt.suptitle('Bivariate Dsitribution', y=1.02)
plt.show()
```

## Multivariate Analysis
Determine the correlation between 2 features . If the 2 features are carrying the same information. When the correlation is very high then there will be redundency if there correlation is over 95 percent. If there is high redundency then then we can decided to drop the feature. 

1. Are 2 features carrying the same information
2. How high is the correlation (95>)
3. Is the feature redundent

```python
df_corr = df_new.drop("Species",axis=1)
correlation_matrix = df_corr.corr()
plt.figure(figsize = (10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f',linewidth=2)
plt.title("Feature Correlation Matrix")
plt.show()
```
