```python
import pandas as pd 
import numpy as np 

```

```python
# Create dataset 
df = pd.DataFrame({
         "Name": ["Scott", "Mike", "Frank", "James", "Maria", "Linda", "Sarah", "Jane"],
         "Gender": ["Male", "Male", "Male", "Male", "Female", "Female", "Female", "Female"],
         "Income": [3000, 4000, np.nan, np.nan, 2000, np.nan, np.nan, 6000],
         "Degree": ["Non-academic", "Non-academic", "Academic", "Academic",
                    "Non-academic", "Non-academic", "Academic", "Academic"]
     })
df
```

```python
df.isna().sum()
```

```python
df.isna().mean() * 100
```

```python
df.groupby("Degree")["Income"].apply(lambda x: x.isna().mean())
```

```python
df_drop = df.dropna()
df_drop
```

```python
len(df_drop)
```

```python
df_drop["Income"].mean()
```

```python
overall_mean = df["Income"].mean()
overall_mean
```

```python
df_mean_impute = df.copy()
df_mean_impute["Income"] = df_mean_impute["Income"].fillna(overall_mean)
df_mean_impute
```

```python
df_mean_impute.groupby("Degree")["Income"].mean()
```

```python

df_group_impute = df.copy()
     
df_group_impute["Income"] = df_group_impute.groupby("Degree")["Income"].transform(
         lambda x: x.fillna(x.mean())
)
     
df_group_impute 

```

```python
df_group_impute.groupby("Degree")["Income"].mean()
```

```python
comparison = pd.DataFrame({
"Original": df["Income"],
"Mean Imputation": df_mean_impute["Income"],
"Group Imputation": df_group_impute["Income"]
})
comparison
```
