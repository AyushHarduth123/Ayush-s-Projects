```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Dataset
data = pd.DataFrame({
    "Temperature": [0.5, 3, 4, 6, 8, 10, 12, 13, 15, 17, 18, 20, 23],
    "GolfPlayers": [0, 0, 0, 1, 1, 7, 9, 8, 7, 10, 10, 10, 10]
})

X = data[["Temperature"]]
y = data["GolfPlayers"]

# Regression Tree
model = DecisionTreeRegressor(
    max_depth=3,   # prevents overfitting
    min_samples_leaf=3
)
model.fit(X, y)

# Visualise tree
plt.figure(figsize=(12, 6))
plot_tree(
    model,
    feature_names=["Temperature"],
    filled=True,
    rounded=True
)
plt.show()
```
