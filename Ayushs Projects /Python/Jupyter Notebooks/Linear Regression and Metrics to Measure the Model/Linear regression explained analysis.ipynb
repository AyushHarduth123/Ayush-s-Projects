## Four assumptions of Linear Regression

1. **Linear Relationship**: it ensures dependent variables changes with independent variables. In terms of the residual vs predicted, there must not be any form of relationship.  
   **Solution**: Polynomial regression or some feature transformation.

2. **Independence of error**: It compares the errors between the actual and predicted. It ensures there is no correlation between the residuals.

3. **Homoscedasticity**: It ensures that the variance of residual remains constant

4. **Normally distributed errors**: ensures valid hypothesis testing. There must be a normal distribution around the zero value.

These assumptions are required to ensure that the regression model produces a valid, reliable and interpretable results.

---

## Residual Diagnostics

This is the process of analyzing the residuals (prediction errors) from the regression model to determine whether the assumptions of regression are satisfied or violated.

1. **Linear Regression** - Residual vs fitted plot  
2. **Homoscedasticity** - Residual vs fitted plot  
3. **Normal distribution of errors** - Histogram or Q - Q Plot (Quantile - Quantile Plot)

```python
df_result= x_test.copy()
df_result['Actual'] = y_test
df_result['Predicted'] = y_pred
df_result['Residuals'] = y_test-y_pred
```

```python
plt.scatter(y_pred,residual)
plt.axhline(y=0, color='red')
plt.xlabel("predicted values")
plt.ylabel("residuals")
plt.title("Scatter plot showing residuals v predicted " )
plt.show()
```

note:

the residual plot checks the linearity and homoscdasicity 

explanation: randomly scattered plot around the zero line 

the model is considered to be a good model:
1. residuals randomly scattered 
2. no clear pattern

there is a problem if
1. there is a curved pattern - there is a relationship but it is not linear 
2. there is a funnel - homoscedasicity exists 
3. there are clusters - it means there are missing values 

### Histo gram of Residuals

The purpose is to check if residuals follows a normal distribution 

Expectation: a bell shaped distribution around 0

If this assumption is not satisfied, this means the distribution is skewed, it means some variables are skewed or heavily tailed


```python
import seaborn as sns 

sns.histplot(residual, kde=True)
plt.title('Distribution of residuals ')
plt.show()
```

#### Quantile plot 

```python
import scipy.stats as stats
stats.probplot(residual,dist="norm",plot = plt)
plt.title('Q-Q Plot')
plt.show()
```

#### Diagnostic Check 



```python
df_residuals = df_result.drop(["Actual","Predicted","Residuals"], axis=1)
```

```python
fig,axes = plt.subplots(4,3,figsize=(20,20))
axes = axes.flatten()

for i, feature in enumerate(df_features):
    axes[i].scatter(df_residuals[feature],residual)
    axes[i].axhline(0,color = 'red')
    axes[i].set_title(f"residual vs {feature}")
```

A good residual diagnostics should ensure the following:

1. Residuals randomly scattered around zero
2. Constant variance
3. Approximately normal distribution around zero
4. No correlation between residuals




### RIDGE AND LASSO REGRESSION 
RIDGE SHRINKS COEFFIECENTS NO NEAR 0 
LASSO REGRESSION SHRINKS COEFFIECENTS TO 0 PERFORMS FEATURE SELECTION 

Used where there is multicolinearity 



```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1.0)
ridge.fit(x_train,y_train)
```

```python
y_pred_ridge = ridge.predict(x_test)
```

```python
mae_ridge = mean_absolute_error(y_test,y_pred_ridge)
print(mae_ridge)
mse_ridge = mean_squared_error(y_test,y_pred_ridge)
print(mse_ridge)

r2_ridge = r2_score(y_test,y_pred_ridge)
print(r2_ridge)
```

## Lasso Regression
