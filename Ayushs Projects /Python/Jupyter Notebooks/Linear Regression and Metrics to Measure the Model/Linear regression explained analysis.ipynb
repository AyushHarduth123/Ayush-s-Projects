```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axes = plt.subplots(1,2,figsize = (12,6))

sns. histplot(df_new['SalePrice'], ax = axes[0], kde = True,bins=30)
plt.title("Dsitribution os Sales Prices")
plt.xlabel("Sales Prices")

axes[1].boxplot(df_new['SalePrice'])

axes[1].boxplot(df_new['SalePrice'])
axes[1].set_title("BoxPlot")
axes[1].set_ylabel("Sale Price")
```

```python
# Target is right skewed, meaning that metrics will be skewed when calculating things like the mean, making it difficult to calculate and make good assumtions on the data 
# The box plot 
# any points outside of the boxplot it is considered as outliers 
# allows you to understand the distribution of the data whe the data is pushed to the top it is considered left skewed and conversly when it is going down 
# apply a log transformation to the dataset to make it normal
df_new['SalePrice_log'] = np.log1p(df_new['SalePrice'])
```

```python
# Fixing the box plot
# why and how it is analysed

fig, axes = plt.subplots(1,2,figsize = (12,6))

sns. histplot(df_new['SalePrice_log'], ax = axes[0], kde = True,bins=30)
plt.title("Dsitribution os Sales Prices")
plt.xlabel("Sales Prices")

axes[1].boxplot(df_new['SalePrice_log'])

axes[1].boxplot(df_new['SalePrice_log'])
axes[1].set_title("BoxPlot of the log transformation ")
axes[1].set_ylabel("Sale Price")
```

## Explaing the difference 
1. in the first plot the regression line will be shifted towards the right as there are more outliers that skews the models results towards the right of the model
2. In order to fix this we carried out a log transformation on the dtaa to balance the distribution of the prices. This now balances the distribution of the values where there are now more balanced distributions of the data.
3. When there is no lineararity between the features then you would have to transform the data or discard it

```python
# Bivariate analysis 
# relationship between each variable and the target 
# remove the sales 
df_columns = df_new.drop(['SalePrice','SalePrice_log'], axis =1)
df_features = []
df_features = df_columns.columns
```

```python
df_features
```

```python
fig, axes = plt.subplots(4,3,figsize = (20,20))
axes = axes.flatten()

for i, feature in enumerate(df_features):
    axes[i].scatter(df_new[feature],df_new['SalePrice_log'], alpha=0.5,s=10)
    sns.regplot(x=df_new[feature], y=df_new['SalePrice_log'], ax = axes[i], scatter = False, lowess=True, color='red')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel("Sale Price")
    axes[i].set_title(f'(feature) v Target')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(4,3,figsize = (20,20))
axes = axes.flatten()

for i, feature in enumerate(df_features):
    axes[i].scatter(df_new[feature], df_new['SalePrice'], alpha=0.5, s=10)
    sns.regplot(x=df_new[feature], y=df_new['SalePrice'],
                ax=axes[i], scatter=False, lowess=True, color='red')
    
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel("Sale Price")
    axes[i].set_title(f'{feature} vs Target')

plt.tight_layout()
plt.show()
```

```python
correlations = df_new.corr()['SalePrice_log'].sort_values(ascending=False)
print(correlations)

plt.figure(figsize = (12,8))
sns.heatmap(df_new.corr(),annot = True, fmt = '.2f', cmap = 'coolwarm', square = True, linewidth=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_new[df_features]
y = df_new['SalePrice_log']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Encoding comes next 
# then Scaling
# ice task: pass the skewed data into the algorithm to see what the performance improvments will be 

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
# MSE 
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test,y_pred)
print(mae)

# we want a small value meaning it is a good model 
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
print(mse)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print(r2)

# high score is a good thing 
```
