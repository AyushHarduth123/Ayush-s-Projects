```python
## MULTI CLASS CLASSIFICATION 
from sklearn.datasets import load_iris
```

```python
# global dTA SET 
# applys gobally and is invoked once 
# random state applys per algorythm
np.random.seed(42)
```

```python
iris = load_iris()
```

```python
x,y = iris.data, iris.target
```

```python
iris.DESCR
```

```python
feature_name = iris.feature_names
target_name = iris.target_names
print(f"Feature Names: {feature_name}")
print(f"Feature Names: {target_name}")
```

```python
x.shape
```

```python
df = pd.DataFrame(x, columns=feature_name)
df["Target"] = y
df.head()
```

```python
df["Species"] = [target_name[i] for i in y]
df.head(70)
```

```python
from sklearn.model_selection import train_test_split 
#stratify ensures that the classes are split evenly in the data for both training and test data 
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42, stratify = y)
```

```python
print(f"\n Training set size: {len(x_train)}")
print(f"\n Test set size: {len(x_test)}")
print(f"\n Training Class Distribution: {np.bincount(y_train)}")
print(f"\n Test Class Distribution: {np.bincount(y_test)}")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
print(f"Original Mean = {x_train.mean(axis = 0).round(2)}")
print(f"Scaled Mean = {x_train_scaled.mean(axis = 0).round(2)}")
```

```python
# have 1 or 2 classifiers including the main one that they as us to do
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.naive_bayes import GaussianNB


```

```python
classifiers = {
    'Logistic Regression': LogisticRegression(
        multi_class = 'multinomial',
        solver='lbfgs',
        max_iter = 1000,
        random_state = 42
    ), 
    'Desicion Tree': DecisionTreeClassifier(
        max_depth = 5,
        random_state = 42
    ),
    'Random Forest': RandomForestClassifier(
        n_estimators = 100,
        max_depth = 5,
        random_state = 42
    ),
    'Gradient Boosting':GradientBoostingClassifier(
        n_estimators = 100,
        max_depth = 3,
        random_state = 42
    ),
    'SVC' : SVC(
        kernel = 'rbf',
        probability = True,
        random_state = 42
    ),
    'KNN': KNeighborsClassifier(
        n_neighbors = 5,
        weights = 'distance'
        
    ),
    'Naive Bayes': GaussianNB()
}
```

```python
from sklearn.metrics import (accuracy_score, classification_report,confusion_matrix, balanced_accuracy_score, f1_score)
```

```python
results = []
predictions = {}
```

```python
for name, clf in classifiers.items():

    #train models 
    #fir all the classifiers 
    clf.fit(x_train_scaled,y_train)

    #predict
    y_pred = clf.predict(x_test_scaled)
    y_pred_prob = clf.predict_proba(x_test_scaled)

    # Metric 
    accuracy = accuracy_score(y_test,y_pred) #performance of model with considering imbance in the class and gives a overall score of the model
    balanced_acc = balanced_accuracy_score(y_test,y_pred)# accounts for the imbalance in the data when calculating the accuracy score 

    # f1 score 
    # f1 score 
    f1_macro = f1_score(y_test, y_pred, average='macro')# calculated the f1 score of the classes then creates a average for the classes 
    f1_weighted = f1_score(y_test, y_pred, average = 'weighted')# considers the weights of each of the classes in the test and training dataset as well as how balanced and inbalanced the classes are and then gives a score 
    
    # Store result 
    results.append({
        'Classifier': name,
        'Accuracy': accuracy,
        'Balanced Accuracy':balanced_acc,
        'F1 (Macro)': f1_macro,
        'F1 (Weighted)': f1_weighted
    })

    ##predictions 
    predictions[name] = y_pred

    print("\n")
    print(f"{name}")
    print(f"Test Accuracy: {accuracy:.4f}")
    #print(f"Classificationn Report: {classification_report(y_test,y_pred)}")

```

```python
print("Result Summary")
results_df = pd.DataFrame(results).sort_values("Accuracy",ascending=False)
results_df
```

```python
predictions
```

```python
df_pred = pd.DataFrame(predictions)
df_pred["Actual"] = y_test
df_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
for name, array in predictions.items():
    print(f"{name} Confusion Matrix")
    ConfusionMatrixDisplay.from_predictions(y_test,array)
    plt.show()
    print("\n")
```

```python
# Classification report 
for name, array in predictions.items():
    print(f"{name} Classification Report")
    print(classification_report(y_test,array))

    print("\n")
```
